```python
#Percorrrer todos os arquivos
#importando bibliotecas
import os
import pandas as pd
import plotly.express as px

list_data = os.listdir("/content/drive/MyDrive/Curso de phyton/Vendas")

#tabela vaiza
table_total = pd.DataFrame()

#Importar base de dados vendas
for arquivo in list_data:
  #se tem "Vendas" no arquivo
  if "Vendas" in arquivo:
    #importar arquivos
    table = pd.read_csv(f"/content/drive/MyDrive/Curso de phyton/Vendas/{arquivo}")
    #resultado armazenando na minha tabela vazia
    table_total = table_total.append(table)

```

```python
#Calcular produto mais vendido (em quantidade)
table_produtcs = table_total.groupby('Produto').sum(numeric_only=True)
table_produtcs = table_produtcs[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False)
display(table_produtcs)


```

```python
#Produto que mais faturou
table_total['Faturamento'] = table_total['Quantidade Vendida'] * table_total['Preco Unitario']

table_faturamento = table_total.groupby('Produto').sum(numeric_only=True)
table_faturamento = table_faturamento[["Quantidade Vendida","Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(table_faturamento)
```

```python
#Loja que mais vendeu
table_lojas = table_total.groupby('Loja').sum(numeric_only=True)
table_lojas = table_lojas[["Quantidade Vendida","Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(table_lojas)
```

```python
#exibindo grafico

grafico = px.bar(table_lojas,x= table_lojas.index,y='Faturamento')
grafico.show()
```
